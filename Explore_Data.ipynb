# Exploring the data

```python
import pickle
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = None
with open('data/data.pickle', 'rb') as f:
    data = pickle.load(f)

# Create a dataframe
df = pd.DataFrame(data)
```

```python
math = df['math score']
reading = df['reading score']
writing = df['writing score']

#plot data
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
plt.suptitle('Scores')
ax1.scatter(math, reading, color='red', label='math x reading')
ax2.scatter(math, writing, color='blue', label='math x writing')
ax3.scatter(reading, writing, color='green', label='reading x writing')
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(math, reading, writing, color='red', label='math x reading x writing')
```
